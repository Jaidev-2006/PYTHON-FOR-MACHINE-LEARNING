```python
import pandas as pd
df=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],
               index=['one','two','three'],
               columns=['a','b','c'])
print(df)
```

```python
#Taking the input from Dataframe and storing in the Excel file
print("Taking the input from Dataframe and storing in the Excel file")
df.to_excel('d:\pandas_to_excel.xlsx',sheet_name='new_sheet_name')
```

```python
#Second Dataframe input to another Excel fild
d=pd.DataFrame([[110,210,310],[12,220,320],[310,320,330]],
              index=['four','five','six'],columns=['a','b','c'])
d.to_excel('d:\pandas_to_excell.xlsx',sheet_name='new_sheet_name')
```

```python
#Merging two Excel files input into third file
x=pd.read_excel('d:\pandas_to_excell.xlsx')
y=pd.read_excel('d:\pandas_to_excell.xlsx')
z=pd.concat([x,y])
z.to_excel('d:\pandas_to_excel3.xlsx')
```

```python
#Sorting the column values
df=z.sort_values(["a"])
print(df)
df.to_excel('d:\pandas_to_excel4.xlsx')
```
